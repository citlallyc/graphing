```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rects1 = plt.bar(index, men, bar_width, alpha=opacity, 
                 color='g', label='Men')

rects2 = plt.bar(index + bar_width, women, bar_width, 
                 alpha=opacity, 
                 color='r',
                 label='Women')

rects3 = plt.bar(index + bar_width + bar_width, unknown, bar_width, 
                 alpha=opacity, 
                 color='y', 
                 label='Unknown')

plt.xlabel('Gender')
plt.ylabel('Subject')
plt.title('Pipeline Data')
plt.xticks(index + bar_width, ('AC', 
           'PS', 
           'TA', 
           'TI', 
           'Offer', 
           'Hired'))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
detail_21 = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 
                       'June', 'Jul', 'Aug', 'Sep', 
                       'Oct', 'Nov', 'Dec'],
             'Hires': [9, 7, 11, 5, 9, 
                    12, 11, 22, 9, 16, 
                    15,16]}
```

```python
det_21 = pd.DataFrame(detail_21)
det_21.plot.line()


sns.barplot(x="Month", y="Hires", data=det_21)
plt.xticks(rotation=1)
plt.title('2021 Hire Details')
plt.xlabel('Month')
plt.ylabel('Hires')
plt.show()


```

```python
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
sns.barplot(x="Month", y="Hires", data=det_21)
show_values(sns.barplot(x="Month", y="Hires", data=det_21))
plt.xticks(rotation=1)
plt.title('2021 Hire Details')
plt.xlabel('Month')
plt.ylabel('Hires')
plt.ylim(0,24)
plt.show()

```
